#### 1. Write a program to take principal amount, rate and interest as input and calculate simple interest.

```python
amount = int(input('Enter the amount : '))                   #Enter the amount
rate = int(input('Enter the rate of interest : '))           #Enter the rate of interest
time = int(input('Enter the year for the interest : '))      #Enter the duration for the rate of interest

#Simple interest = principle amount * rate of interest * time interval  ----> Formula to calculate Simple interest

S = (amount*rate*time)/100                                   #----> Formula to calculate Simple interest

print('The simple interest = ', S)
```

#### 3. Write a program to take a list of numbers as input (in a single line) and print the sum, avg, max and min elements.

```python
# Take a list of numbers as input from the user (comma-separated)
input_string = input("Enter a list of numbers (Can be INT or FLOAT) (comma-separated): ")

# Convert the input string to a list of numbers
number_list = [float(num) for num in input_string.split(',')]

# Calculate the sum, average, maximum, and minimum elements
total_sum = sum(number_list)
average = total_sum / len(number_list)
max_element = max(number_list)
min_element = min(number_list)

# Print the results
print("List of numbers which you entered:", number_list)
print("Sum of all Numbers which you entered:", total_sum)
print("Average of all Numbers which you entered:", average)
print("Maximum Number from Numbers which you entered:", max_element)
print("Minimum Number from Numbers which you entered:", min_element)

```

#### 4. Write a program to take the marks of 3 students in five subjects each as a list (one list for each input) as input and print the average score of each student..

```python
#Define the function to Calculate average so we can use it later
def calculate_average(marks):
    return sum(marks) / len(marks)

# Initialize empty lists to store marks for each student and each subject
students = []

# Take input for each student
for i in range(3):
    student_marks = []
    print(f"\nEnter marks for Student {i+1}:")
    hindi = float(input("Hindi: "))
    english = float(input("English: "))
    maths = float(input("Maths: "))
    physics = float(input("Physics: "))
    chemistry = float(input("Chemistry: "))

    student_marks.extend([hindi, english, maths, physics, chemistry])
    students.append(student_marks)

# Print the entered marks for each subject for each student
for i, student in enumerate(students, start=1):
    print(f"\nEntered Marks for Student {i}: {student}")

# Calculate average scores for each student
averages = [calculate_average(student) for student in students]

# Print the average scores for each student
for i, avg in enumerate(averages, start=1):
    print(f"\nAverage Score for Student {i}: {avg:.2f}")
```

#### 5. Write a program to take a list as input containing results of 10 coin tosses as 'H' or 'T'. Jon gets a point for each 'T' and Tom gets a point for each 'H'. print the points recieved by both.

```python
# Take a list as input containing results of coin tosses
result_coin_toss = input("Enter the results of coin tosses (H/T, separated by commas): ")

# Split the input string into a list of individual coin toss results
results_list = result_coin_toss.split(',')

# Check if exactly 10 results are entered
if len(results_list) != 10:
    print("More than 10 results entered. Please enter exactly 10 results.")
else:
    # Count the points for Jon and Tom
    jon_points = results_list.count('T')
    tom_points = results_list.count('H')

    # Print the points received by Jon and Tom
    print("Points received by Jon:", jon_points)
    print("Points received by Tom:", tom_points)


```

#### 7. Take a list as input in a single line containing 10 numbers as marks out of 100. The even positions in the list are marks of Tom and the odd positions are the marks of Jon. Calculate the average marks of both.

```python
# Take a list as input containing 10 numbers (marks out of 100)
marks_list = list(map(float, input("Enter 10 marks (comma-separated, out of 100): ").split(',')))

# Check if exactly 10 marks are entered
if len(marks_list) != 10:
    print("Wrong input. Please enter exactly 10 marks.")
else:
    # Print the user input
    print("User input:", marks_list)

# Extract marks of Tom (even positions) and Jon (odd positions) from the list
tom_marks = marks_list[::2]
jon_marks = marks_list[1::2]

# Calculate the average marks of Tom and Jon
tom_average = sum(tom_marks) / len(tom_marks)
jon_average = sum(jon_marks) / len(jon_marks)

# Print the average marks of Tom and Jon
print("Average marks of Tom:", tom_average)
print("Average marks of Jon:", jon_average)
```

#### 8. Take a string and a character as input. Print the result having the character between each word of the given string.

```python
# Take a string as input
input_string = input("Enter a string: ")

print("The string which you have entered is ",input_string)

# Take a character as input
char = input("Enter a character: ")
print(f"Let's count occurrence of '{char}' in the string which you have entered: '{input_string}'")

# Count the occurrence of the input character in the input string
char_count = input_string.count(char)

# Print the result
print(f"Occurrence of character '{char}' in the string is:", char_count)


# Count the occurrence of each character in the input string
char_count = {}
for c in input_string:
    char_count[c] = char_count.get(c, 0) + 1

# Print the result in the required format (as given in assignment example)
print("Occurrence of all characters in", input_string, "is:")
print(char_count)

```

#### 9. Create a dictionary having three items. Key of each item is the name of a student, value associated with each key is a list of marks in five subjects. Given this dictionary print the average score of each student.

```python
# Create a dictionary with three students and their marks in five subjects
student_marks = {
    'John': [80, 75, 85, 90, 88],
    'Bipasha': [70, 82, 78, 88, 95],
    'Dino': [90, 80, 85, 92, 75]
}

# Calculate and print the average score of each student
for student, marks in student_marks.items():
    average_score = sum(marks) / len(marks)
    print(f"Average score of {student}: {average_score:.2f}")
```

#### 10. Take a string as input(may contain repetitive words), print a list consisting of unique words from the given string.

```python
# Taking a string as input
input_string = input("Enter a string: ")
print(f"you have entered string as  '{input_string}'")

# Splitting the input string into individual words
words = input_string.split()
print(f"your '{input_string}' will be split into words as '{words}'")

# Creating a set from the words to remove duplicates
unique_words = set(words)
print(f"your '{input_string}' has these unique words '{unique_words}'")

# Convert the set back to a list to get unique words
unique_words_list_input = list(unique_words)

# Print the list of unique words
print("List of unique words from the input string:")
print(unique_words_list_input)
```
